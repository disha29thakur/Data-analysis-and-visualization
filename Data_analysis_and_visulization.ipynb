**Survival Rates:**
In survival rate analysis these are the findings :


1.   The total survival rate is 38.38%.
2.   Female have more survival rate compared to male.
     Female survival rate - 74.20%
     Male suvival rate - 18.89%
2.   Survival rate as per passenger class is:
     1 class - 62.96%
     2 class - 47.28%
     3 class - 24.24%





```python
# Overall survival rate

total_survived = df['Survived'].sum()
total_passengers = df['PassengerId'].count()

overall_survival_rate = (total_survived/total_passengers)*100
print(overall_survival_rate)
```

```python
# Overall survival rate by gender
gender_survival_rate = df.groupby('Sex')['Survived'].mean() * 100
print(gender_survival_rate)
```

```python
# Survival rate based on each passenger class
passenger_class_srate = df.groupby('Pclass')['Survived'].mean()*100
passenger_class_srate= round(passenger_class_srate,2)
print(passenger_class_srate )
```

```python
plt.figure(figsize = (7,6))
plt.bar(['1','2','3'],passenger_class_srate,color=['pink','yellow','purple'],width=0.2)
plt.title('Bar graph of survival rate')
plt.xlabel('Class')
plt.ylabel('Number of people')

plt.title('Bar graph of survival rate')
plt.xlabel('Class')
plt.ylabel('Number of people')

# Annotate the bars with the count
for i, count in enumerate(passenger_class_srate):
    plt.text(i, count, str(count), ha='center',va='bottom')
plt.show()
```

Outlier detection : Finding very extreme and low values which can impact the perfomance of the model.


*   To detect outlier we have used IQR method.
*   Those data points which are less than lower bound and higher than the upper bound are potential outliers.



```python
# Identifying outlier in Fare column
Q1 = 7.910400 #25%
Q3 = 31.000000  #75%
IQR = Q3-Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['Fare'] < lower_bound) | (df['Fare'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")
print(f"Maximum of outliers: {outliers['Fare'].max()}")
print(f"Minimum of outliers: {outliers['Fare'].min()}")
print(outliers[['Fare']])
```

Additional analysis

We constructed bins for the age column and performed an age-survival analysis. It indicates which group of individuals survived or not.





```python
bins = [0,20,40,60,80]
labels = ['young', 'middle-aged', 'old', 'very-old']
df['age_bins'] = pd.cut(df['Age'], bins, labels=labels)
```

```python
df.groupby('age_bins')['Survived'].count()
```

```python
survival_counts=df.groupby('age_bins')['Survived'].value_counts()
print(survival_counts)
```
